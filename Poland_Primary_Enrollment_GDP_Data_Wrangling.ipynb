**Filtering down data**

```python
#filter e down to GDP inidcators
e_filtered = e[e['Indicator Name'].str.contains('GDP')]
#showcase unique indicators
e_filtered['Indicator Name'].unique()
```

```python
#filter down to specific indicator selected for GDP
e_filtered = e_filtered[e_filtered['Indicator Name'] == 'GDP (current US$)']
```

```python
#show all countries listed in the set
e_filtered['Country Name'].unique()
```

```python
#filter e to the specific selected country, Poland
poland = e_filtered[e_filtered['Country Name'].str.contains('Poland')]
#verify the set has what is needed for GDP
poland.head()
```

```python
#begin filtering G to the selected country, Brazil
poland_g = g[g['Country Name'] == 'Poland']
```

**Restrucuring data**

```python
#create a variable holding the years we are wanting to look at
years = [str(y) for y in range(2001, 2021)]
#create a variable with the columns we want to keep
select_id = ['Country Name','Country Code','Indicator Name','Indicator Code']
#combine the variables as the variable, select
select = select_id+years

#use the select variable to structure the data to what we want to utilize
poland_structured = poland[select]
#show the structured data to confirm it is what we need
poland_structured.head()
```

```python
#utlize ment to restructure the format of the data so we have years as a value
#this allows us to join the other set on years, once cleaned
poland_melt = poland_structured.melt(
    id_vars=select_id,
    value_vars = years,
    var_name = 'Year',
    value_name = 'GDP'
    )
```

```python
#show the set change to make sure it looks correct
poland_melt.head()
```

```python
#utilize the same method to melt the data into a format we can use
poland_g_melt = poland_g.melt(
    id_vars = select_id,#we can use the same variables for selection as we did prior as they match between sets
    value_vars = years,
    var_name = 'Year',
    value_name = 'Primary education'
)
```

```python
poland_g_melt.head()
```

**Merging the data**

```python
#join the seats as poland_merged on year
merged = pd.merge(poland_g_melt , poland_melt, on ='Year', how = 'inner')
```

```python
merged.head()
```

```python
#drop unnecessary columns we do not need for the primary set
poland_final = merged.drop(['Country Name_y',	'Country Code_y',	'Indicator Name_y',	'Indicator Code_y',	'Indicator Name_x',	'Indicator Code_x','Country Name_x',	'Country Code_x'] , axis = 1)
#rename column to match. Include  '_' for SQL DB
poland = poland_final.rename(columns = {'Primary education':'Primary_Enrollment'})
```

**Final set review**

```python
#show the final product to varify it is what we are looking for
poland
```

```python
#use the interpolate method to fill missing values for primary education
poland['Primary_Enrollment'] = poland['Primary_Enrollment'].interpolate()
poland
```

```python
#plot the data as a scatter chart to visualize 

#set the size of the chart
plt.figure(figsize=(12,7))
#create the chart itself
scatter = plt.scatter(x=poland['Year'], #setting the x axis as year
                     y=poland['Primary_Enrollment'], #setting the y axis to enrollment
                     c=poland['GDP'], #setting colormapint to GDP
                     cmap='coolwarm')#setting coolwarm to show the GDP changes
#create labels 
plt.xlabel("Year")
plt.ylabel("Primary Education")
plt.title("Poland Primary Education Enrollment by GDP")
plt.colorbar(scatter, label='GDP (in billions USD)')  
#show the chart
plt.show()
```

```python
#create db connection 
conn = sqlite3.connect(r'C:\Users\buckm\PY\CLASS_FILES\MODULE_8\DATA\poland.db')
#create object to execute SQL
cur = conn.cursor()
```
